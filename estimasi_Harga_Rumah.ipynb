# Harga Rumah Berdasarkan Grade

```python
print(df.columns)
```

```python
grade = df.groupby('grade').count()[['price']].sort_values(by='price',ascending=True).reset_index()
grade = grade.rename(columns={'price':'numberOfprice'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=grade['grade'], y=grade['numberOfprice'], color='royalblue')
plt.xticks(rotation=60)
```

# Kamar Tidur


```python
jumlahkamar= df.groupby('bedrooms').count()[['grade']].sort_values(by='grade',ascending=True).reset_index()
jumlahkamar = jumlahkamar.rename(columns={'grade':'count'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=jumlahkamar['bedrooms'], y=jumlahkamar['count'], color='royalblue')
plt.xticks(rotation=60)
```

# Kamar Mandi

```python
jumlahkamarmandi= df.groupby('bathrooms').count()[['grade']].sort_values(by='grade',ascending=True).reset_index()
jumlahkamarmandi = jumlahkamarmandi.rename(columns={'grade':'count'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=jumlahkamarmandi['bathrooms'], y=jumlahkamarmandi['count'], color='royalblue')
plt.xticks(rotation=60)
```

# seleksi fitur

```python
features = ['bedrooms','bathrooms','floors','grade','condition','view','waterfront']
x = df[features]
y = df['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python

input_data = np.array([[5,3,2,6,3,2,1]])

prediction = lr.predict(input_data)
print('Estimasi Harga Rumah :', prediction)
```

# save model

```python
import pickle

filename = 'estimasi_Harga_Rumah.sav'
pickle.dump(lr,open(filename,'wb'))
```
